```python
#  data
content = """sky,airtemp,humidity,wind,water,forecast,enjoysport
sunny,warm,normal,strong,warm,same,yes
sunny,warm,high,strong,warm,same,yes
rainy,cold,high,strong,warm,change,no
sunny,warm,highstrong,cool,change,yes
"""
with open("sport.csv","w") as f:
    f.write(content) 


import numpy as np
import pandas as pd

# Load data
data = pd.read_csv("sport.csv")
print(data)

# Extract concepts and target
concepts = np.array(data)[:, :-1]
target = np.array(data)[:, -1]
print("Concepts:\n", concepts)
print("Target:\n", target)

# Candidate Elimination Algorithm
def learn(concepts, target):
    # Initialize Specific and General Hypotheses
    specific_h = concepts[0].copy()
    print("\nInitialization of specific_h and general_h")
    print("\nSpecific Boundary: ", specific_h)
    
    # Initialize generic hypothesis boundary with '?' for each attribute
    general_h = [["?" for i in range(len(specific_h))] for i in range(len(specific_h))]
    print("\nGeneric Boundary: ", general_h)
    
    for i, h in enumerate(concepts):
        print("\nInstance", i+1, "is ", h)
        
        if target[i] == "yes":  # Positive instance
            print("Instance is Positive")
            for x in range(len(specific_h)):
                if h[x] != specific_h[x]:
                    specific_h[x] = '?'  # Adjust specific boundary
                    general_h[x][x] = '?'  # Adjust general boundary to be more general

        if target[i] == "no":  # Negative instance
            print("Instance is Negative")
            for x in range(len(specific_h)):
                if h[x] != specific_h[x]:
                    general_h[x][x] = specific_h[x]  # Adjust general boundary to be more specific
                else:
                    general_h[x][x] = "?"  # Keep general boundary as is

        print("Specific Boundary after instance ", i+1, " is ", specific_h)
        print("Generic Boundary after instance ", i+1, " is ", general_h)

    # Remove the '?' combinations that are no longer valid
    indices = [i for i, val in enumerate(general_h) if val == ['?', '?', '?', '?', '?', '?']]
    for i in indices:
        general_h.remove(['?', '?', '?', '?', '?', '?'])

    return specific_h, general_h

# Running the learn function
s_final, g_final = learn(concepts, target)

# Display final Specific and General Hypotheses
print("\nFinal Specific Hypothesis: ", s_final, sep="\n")
print("\nFinal General Hypothesis: ", g_final, sep="\n")
```
